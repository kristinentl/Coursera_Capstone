<h3>Task 1:Scraping the Wikipedia - postal codes of Canada</h3>

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

<h3>html table to Pandas DataFrame for cleaning & preprocessing</h3>

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

<h3>Data preprocessing & cleaning</h3>

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

```python
# Shape of data frame
df2.shape
```

<h3>Importing the csv - latitudes & longitudes</h3>

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

<h3>Latitudes & Longitudes for various neighbourhoods in Canada</h3>

```python
df3 = pd.merge(df2,lat_lon,on='Postal Code')
df3.head()
```

<h2>Taks 2: Clustering & Plotting of the neighbourhoods of Canada</h2>

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

<h3>Visualizing all the Neighbourhoods - Folium</h3>

```python
torontomap = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.Circle(
    [lat,lng],
    radius=5,
    popup=label,
    color='yellow',
    fill=True,
    fill_color='blue',
    fill_opacity=0.9,
    parse_html=False).add_to(torontomap)
torontomap
```

<h3>KMeans clustering for the clsutering of the neighbourhoods</h3>

```python
#Markdown and LaTeX:  𝛼2
k=5
toronto_clustering = df4.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=8)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.9).add_to(map_clusters)
       
map_clusters
```
